```python
import numpy as np
import pandas as pd
```

```python
from numpy.random import randn
```

```python
np.random.seed(101)
```

```python
df = pd.DataFrame(randn(5, 4), ['A', 'B', 'C', 'D', 'E'], ['W', 'X', 'Y', 'Z'])
df
```

```python
#You can use conditional statements to apply to every
#element of the df
df > 0
```

```python
#Ths line of code will keep all entries that are greater than 0
#and replace all other entries with NaNs / nulls
df[df > 0]
```

```python
#You can make an entire new column with just one value
df['new'] = 'new'
df
```

```python
#This line throws an error because you can't
#use the ">" operater with strings
df > 0
```

```python
#Get the original dataframe back
df.drop('new', axis = 1, inplace = True)
df
```

```python
df['W'] > 0
```

```python
#Access the W column
df['W']
```

```python
#Return the subdataframe that has W entries that are greater
#than 0
df[df['W'] > 0]
```

```python
#Return the X column, but only the entries that have rows
#Where the W column entries are greater than 0.
df[df['W'] > 0]['X']
```

```python
#The following snippets of code do the same thing:
df[df['W'] > 0][['Y', 'X']]
```

```python
boolser = df['W'] > 0
result = df[boolser]
mycols = ['Y', 'X']
result[mycols]
```

```python
#You can use & or | for more complex boolean conditions
#Note: "and" or "or" will raise an error
df[(df['W'] > 0) and (df['Y'] > 1)]
```

```python
df[(df['W'] > 0) & (df['Y'] > 1)]
```

```python
df[(df['W'] > 0) | (df['Y'] > 1)]
```

```python
df.reset_index()#inplace = False
```

```python
newind = 'CA NY WY OR CO'.split()
df['States'] = newind
df
```
