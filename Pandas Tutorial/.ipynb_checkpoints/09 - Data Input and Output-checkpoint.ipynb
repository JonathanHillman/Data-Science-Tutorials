```python
#This is how you read tables from html, it usually returns a list of dataframes
#To view a specific dataframe, you'll have to index the list
data = pd.read_html('http://www.fdic.gov/bank/individual/failed/banklist.html')
type(data)
```

```python
df = data[0]
df.head()
```

```python
#You should generally use a different driver for different sql databases,
#But here is a pandas example
from sqlalchemy import create_engine
engine = create_engine('sqlite:///:memory:')
```

```python
df.to_sql('my_table', engine)
```

```python
sqldf = pd.read_sql('my_table', engine)
sqldf
```
