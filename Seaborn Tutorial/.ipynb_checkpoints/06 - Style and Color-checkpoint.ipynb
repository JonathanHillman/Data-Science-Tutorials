```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
#You can use sns.set_style to set the style of a plot before you create it
#sns.set_style("ticks") gives tick marks to the graph
#sns.set_style("white") makes it blank
#sns.set_style("whitegrid")
sns.set_style("darkgrid") #Gives a dark backgroudn to the graph with white lines
sns.countplot(x = 'sex', data = tips)
```

```python
#You can use matplotlib to change some graphics with seaborn
plt.figure(figsize = (12, 3))
sns.countplot(x = 'sex', data = tips)
```

```python
#Seaborn can set the context of a graph so you can change it for different settings.
#Popular contexts include: notebook, paper, talk and poster
#You can always set additional attributes with another parameter call.
#The default context is notebook
sns.set_context('poster', font_scale = 3)
sns.countplot(x = 'sex', data = tips)
```

```python
#This context is applied for the rest of the figures in the document,
#So we need to set it back to notebook
#There are plenty of palettes to use in sns, just google them to find them
sns.set_context("notebook")
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, 
           hue = 'sex', palette = 'seismic')
```
