```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
#sns.lmplot makes a Linear Model Plot with correspong x and y axis.
#You can also use matplotlib style parameters such as hue, markers,
#The height parameter specifies the height in inches, and the aspect
#parameter specifies the width / height ratio. So, aspect * height = width.
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, 
           hue = 'sex', markers = ['o', 'v'])
```

```python
#You can do a FacetGrid with the col and row parameters
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, row = 'time', col = 'sex')
```

```python
#seaborn aspect ratios can also be adjusted
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, 
           row = 'time', col = 'sex',
          aspect = 0.6, height = 8)
```
