```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
#seaborn comes with a few built in datasets
tips = sns.load_dataset('tips')
tips.head()
```

```python
#This will make a distribution plot of the total bills of the tips df
#The kde is the kernel density estimation, and is represented by the smooth line
#The default value of kde is True
#You can change the number of bins with the bins argument.
sns.distplot(tips['total_bill'], kde = True, bins = 30)
```

```python
#seaborn offers bivariate distribution plots
#The default "kind" parameter is scatter, which is dots.
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'reg')
```

```python
#pairplot will make a jointplot for every pair of features in a dataframe
#For the pairs that are the same, ex. total_bill vs total_bill, there will
#be a histogram instead.
#The hue parameter can be used like so:
#The pallete changes the color
sns.pairplot(tips, hue = 'sex', palette = 'coolwarm')
```

```python
#rugplots and histograms are similar, but rugplots have
#very small bins and uniform height among the graph
sns.rugplot(tips['total_bill'])
```
