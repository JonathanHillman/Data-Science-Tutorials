```python
import seaborn as sns
import numpy as np
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
#sns.barplot will call the estimator function on the y column for each group
#in the x call. 
sns.barplot(x = 'sex', y = 'total_bill', data = tips, estimator = np.mean)
```

```python
#A count plot is the same as a barplot, but with the
#y axis has been chosen as the count
sns.countplot(x = 'sex', data = tips)
```

```python
#A boxplot is also known as a box and whisker plot.
#The points are outliers, and the horizontal lines are the minimum,
#Q1, median, Q3, and max non outlier values.
#You can also apply a value to the hue parameter.
sns.boxplot(x = 'day', y = 'total_bill', data = tips, hue = 'smoker')
```

```python
#A violin plot uses kernel density estimation (kde) to show the distribution
#of the data. This is similar to a boxplot.
#You can also set a hue argument, but if you want the hues to be split,
#Then you can set the split parameter to true
#This type of plot is typically more complicated and harder to read
sns.violinplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex', split = True)
```

```python
#A strip plot is similar to a violin plot or boxplot.
#A problem with a strip plot is that the points may overlap.
#The jitter parameter adds noise to the points to separate them
#Like before, you can also call hue and split
sns.stripplot(x = 'day', y = 'total_bill', data = tips, jitter = True, hue = 'sex', split = True)
```

```python
#You can combine the strip plot and violin plot with a swarm plot
#One drawback of using a swarm plot is that they dont scale very well with large numbers.
#It also uses a lot of computation, so try not to use swarm plots with large datasets
sns.swarmplot(x = 'day', y = 'total_bill', data = tips)
```

```python
#You can combine the plots like so:
sns.violinplot(x = 'day', y = 'total_bill', data = tips)
sns.swarmplot(x = 'day', y = 'total_bill', data = tips, color = 'black')
```
