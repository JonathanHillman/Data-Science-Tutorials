```python
import matplotlib.pyplot as plt
```

```python
#This allows us to view matplotlib graphs in the notebook
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
#There are two ways to create these plots: The functional way,
#and the object oriented way
#Here is the functional way:
plt.plot(x, y)
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Title")
plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

```python
#Object Oriented Method
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y)
axes.set_xlabel("X label")
axes.set_ylabel("Y label")
axes.set_title("Title")
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x, y)
axes2.plot(y, x)

axes1.set_title('LARGER PLOT')
axes2.set_title('SMALLER PLOT')
```

# Summary:

To use the Object Oriented method, you first need to make a figure object, fig,  with the plt.figure() function.

Then, you can add multiple sets of axes to that figure. Using the fig.add_axes() command. 

Then, you can add data, labels, titles, and more to each respective set of axises with axis.plot(), axis.set_xlabel() & axis.set_ylabel(), and axis.set_title() and more respectively.
