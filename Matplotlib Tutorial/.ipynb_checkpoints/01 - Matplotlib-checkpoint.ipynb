```python
import matplotlib.pyplot as plt
```

```python
#This allows us to view matplotlib graphs in the notebook
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
#There are two ways to create these plots: The functional way,
#and the object oriented way
#Here is the functional way:
plt.plot(x, y)
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Title")
plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

```python
#Object Oriented Method
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y)
axes.set_xlabel("X label")
axes.set_ylabel("Y label")
axes.set_title("Title")
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x, y)
axes2.plot(y, x)

axes1.set_title('LARGER PLOT')
axes2.set_title('SMALLER PLOT')
```

# Summary:

To use the Object Oriented method, you first need to make a figure object, fig,  with the plt.figure() function.

Then, you can add multiple sets of axes to that figure. Using the fig.add_axes() command. 

Then, you can add data, labels, titles, and more to each respective set of axises with axis.plot(), axis.set_xlabel() & axis.set_ylabel(), and axis.set_title() and more respectively.

```python
#With the plt.subplots method, we dont need to specify different
#axes variables. Instead, we can just return a list of axes objects
fig, axes = plt.subplots(nrows = 1, ncols = 2)

for current_ax in axes:
    current_ax.plot(x, y)
```

```python
#axes is now an array
axes
```

```python
#This is the same thing. You can index the list specifically
#Or loop through it
fig, axes = plt.subplots(nrows = 1, ncols = 2)

axes[0].plot(x, y)
axes[0].set_title("First Title")

axes[1].plot(x, y)
axes[1].set_title("Second Title")

plt.tight_layout()
```

# Figure Size and DPI

```python
#Make the plot 3 inches by 2 inches
fig = plt.figure(figsize = (3, 2))

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
```

```python
#Make the plot 2 inches by 2 inches, and also add subplots
fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize = (8, 2))

axes[0].plot(x, y)
axes[1].plot(y, x)

#include this for prettier plots
plt.tight_layout()
```

```python
fig.savefig("my_picture.jpg", dpi = 200)
```

```python
#To create a legend, use the ax.legend() function. Then, for ax.legend()
#to create a legend, there must be a label parameter for every ax.plot call.

fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x ** 2, label = 'X squared')
ax.plot(x, x ** 3, label = 'X cubed')

#loc = 0 will try to find the best legend place possible. Its confusing how this
#works, but using loc = n for some other integer n will position the legend elsewhere.
#See the documentation for more details
#You can also add a tuple for a more specific location
#Try ax.legend(loc = (0.1, 0.1))
ax.legend(loc = 0)
```
